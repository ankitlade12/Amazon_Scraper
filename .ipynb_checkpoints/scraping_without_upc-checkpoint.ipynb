```python
#product name
cookie={}
def Searchasin(asin):
    url = 'https://www.amazon.com/dp/'+asin
    print(url)
    page = requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return 'Error'
```

```python
#function
def Searchreviews(review_link):
    url = 'https://www.amazon.com'+review_link
    print(url)
    page = requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return 'Error'
```

```python
from bs4 import BeautifulSoup
product_names = []
url_list = []
data_asin = []
for p in range(1,10):
    url = ('https://www.amazon.com/s?k='+search_query+'&page={}'.format(p))
    url_list.append(url)
    for page in url_list[-1:]:
        #page = url_list[-1:]
        #print(page)
        response = requests.get(page,cookies=cookie,headers=header)
        print(response) 
        soup = BeautifulSoup(response.content)
        #print(soup)
        for i in soup.findAll('div',{'class':'sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col sg-col-4-of-20'}):
            data_asin.append(i['data-asin'])
len(data_asin)
#print(data_asin)
```

```python
link = []
for i in range(len(data_asin)):
    response = Searchasin(data_asin[i])
    soup = BeautifulSoup(response.content)
    for i in soup.findAll('a',{'data-hook':'see-all-reviews-link-foot'}):
        link.append(i['href'])
print(link)
```

```python
final_list = []
for i in range(len(data_asin)):
    response = Searchasin(data_asin[i])
    soup = BeautifulSoup(response.content)
    product_name = soup.find('span',{'class':'a-size-large product-title-word-break'}).text.strip()
    try:
        product_rating = soup.findAll('span',{'class':'a-icon-alt'})[0].text.replace(' out of 5 stars','').replace('Previous page of related Sponsored Products','').strip()
    except AttributeError:
        product_rating = ''
    except IndexError:
        product_rating = ''
    try:
        product_price = soup.findAll('table',{'class':'a-lineitem'})[0].tr.span.text[1:]
    except IndexError:
        product_price = ''
    try:
        product_review = soup.findAll('span',{'id':'acrCustomerReviewText'})[0].text.replace('ratings','').strip()
    except IndexError:
        product_review = ''
    try:
        brand = soup.find('a',{'id':'bylineInfo'}).text.strip()
    except IndexError:
        brand = ''
    except AttributeError:
        brand = ''
    
        
    final=[product_name,product_rating,product_price,product_review,brand]
    final_list.append(final)
```

```python
#reviews, profile name and ratings
profile_name = []
ratings = []
date = []
#tag = []

for j in range(len(link)):
    for k in range(1,2):
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        soup = BeautifulSoup(response.content)
        #soup1 = soup.findAll('div',{'class':'a-section review aok-relative'})
        #soup1 = BeautifulSoup(response.content)
        for l in soup.findAll('span',{'class':'a-profile-name'})[2:]:
            profile_name.append(l.text)
            for i in soup.findAll('i',{'data-hook':'review-star-rating'}):
                ratings.append(i.text)
                for m in soup.findAll('span',{'data-hook':'review-date'}):
                    date.append(m.text)   
```

```python
rev = {'name':profile_name,'ratings':ratings,'date':date}
review_data = pd.DataFrame.from_dict(rev, orient='index')
review_data = review_data.transpose()
pd.set_option('max_colwidth',800)
```

```python
review_data['geography'] = 'USA'
```

```python
#review_data.drop_duplicates(subset=['name'],inplace=True)
review_data.dropna(inplace=True)
```

```python
final = pd.concat([review_data,df1,df2,df3,df4], axis=1)
final.dropna(inplace=True)
final.head()
```

```python
#user ratings cleaning
def star_clean(df_col):
    new_column = []
    for i in df_col.map(lambda x: x.replace('out of 5 stars', '')):
        x = np.array(i, dtype= object)
        new_column.append(x) 
    return new_column
```

```python
final['ratings'] = star_clean(final.ratings)
final.head()
```

```python
def date_conv(d):
    d = str(d)
    x = re.search("on", d)
    d = d[int(x.end()+1):]
    return d
```

```python
final['date'] = final['date'].apply(date_conv)
final.head()
```

```python
def customerid_clean(df_col):
    new_column = []
    for i in df_col.map(lambda x: x.replace('/gp/customer-reviews/', '')[0:14]):
        x = np.array(i, dtype= object)
        new_column.append(x) 
    return new_column
```

```python
final['customer_id'] = customerid_clean(final.links)
final.head()
```

```python
def productid_clean(df_col):
    new_column = []
    for i in df_col.map(lambda x: x.replace(' ','')[-10:]):
        x = np.array(i, dtype= object)
        new_column.append(x) 
    return new_column
```

```python
final['product_id'] = productid_clean(final.links)
final.head()
```

```python
def userid_clean(df_col):
    new_column = []
    for i in df_col.map(lambda x: x.replace('/ref=cm_cr_arp_d_gw_btm?ie=UTF8', '')[26:]):
        x = np.array(i, dtype= object)
        new_column.append(x) 
    return new_column
```

```python
final['user_id'] = userid_clean(final.userid_link)
final.head()
```

```python
#cleaning reviews
final['reviews_clean'] = final['reviews'].apply(lambda x: " ".join(word.lower() for word in x.split()))
final['tags_clean'] = final['tags'].apply(lambda x: " ".join(word.lower() for word in x.split()))
final.head()
```

```python
final.drop(['reviews','tags','links','userid_link'],axis=1,inplace=True)
final.head()
```
