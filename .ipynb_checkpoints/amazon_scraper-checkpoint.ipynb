## Amazon Scraper

```python
#libraries
import re
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36','Referer':'https://www.amazon.com/'}
```

```python
cookie={}
def Searchasin(asin):
    url = 'https://www.amazon.com/dp/'+asin
    page = requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return 'Error'
```

```python
def Searchreviews(review_link):
    url = 'https://www.amazon.com'+review_link
    page = requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return 'Error'
```

```python
#Page number can be changed
data_asin = []
for page in range(1,2):
    url = ('https://www.amazon.com/s?k='+search_query+'&page={}'.format(page))
    response = requests.get(url,cookies=cookie,headers=header) 
    soup = BeautifulSoup(response.content)
    try:
        for i in soup.findAll('div',{'class':'sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col sg-col-4-of-20'}):
            data_asin.append(i['data-asin'])
    except:
        ''
    try:
        for i in soup.findAll('div',{'class':'s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col sg-col-12-of-16'}):
            data_asin.append(i['data-asin'])
    except:
        ''
len(data_asin)
```

```python
link = []
for i in range(len(data_asin)):
    response = Searchasin(data_asin[i])
    soup = BeautifulSoup(response.content)
    for i in soup.findAll('a',{'data-hook':'see-all-reviews-link-foot'}):
        link.append(i['href'])
```

```python
product_list = []
for i in range(len(data_asin)):
    response = Searchasin(data_asin[i])
    soup = BeautifulSoup(response.content)
    try:
        product_name = soup.find('span',{'class':'a-size-large product-title-word-break'}).text.strip()
    except AttributeError:
        product_name = ''
    try:
        product_rating = soup.findAll('span',{'class':'a-icon-alt'})[0].text.replace(' out of 5 stars','').replace('Previous page of related Sponsored Products','').strip()
    except AttributeError:
        product_rating = ''
    except IndexError:
        product_rating = ''
    try:
        product_price = soup.findAll('table',{'class':'a-lineitem'})[0].tr.span.text[1:]
    except IndexError:
        product_price = ''
    try:
        product_review = soup.findAll('span',{'id':'acrCustomerReviewText'})[0].text.replace('ratings','').strip()
    except IndexError:
        product_review = ''
    try:
        brand = soup.find('a',{'id':'bylineInfo'}).text.strip()
    except IndexError:
        brand = ''
    except AttributeError:
        brand = ''
    
        
    product=[product_name,product_rating,product_price,product_review,brand]
    product_list.append(product)
```

```python
product_scraper = pd.DataFrame(product_list,columns=['name','rating','price','review','brand'])
product_scraper['link'] = 'amazon'
product_scraper['product_id'] = data_asin
```

```python
#reviews, profile name and ratings
profile_name = []
ratings = []
date = []

for j in range(len(link)):
    for k in range(1,2):
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        soup = BeautifulSoup(response.content)
        for l in soup.findAll('span',{'class':'a-profile-name'})[2:]:
            profile_name.append(l.text)
            for i in soup.findAll('i',{'data-hook':'review-star-rating'}):
                ratings.append(i.text)
                for m in soup.findAll('span',{'data-hook':'review-date'}):
                    date.append(m.text)   
```

```python
#review_id_link
review_id_link = ''
for j in range(len(link)):
    for k in range(1,2):
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        soup = BeautifulSoup(response.content) 
        for link1 in soup.findAll('a',{'data-hook':'review-title'}):
            review_id_link += link1.get('href')+ 'new_line'
            review_id_link = review_id_link.strip('\n')
```

```python
#review
review = ''
for j in range(len(link)):
    for k in range(1,2):
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        soup = BeautifulSoup(response.content) 
        for link2 in soup.findAll('span',{'class':'a-size-base review-text review-text-content'}):
            review += link2.get_text()+'new_line'
            review = review.strip('\n')
```

```python
#review_tag
review_tag = ''
for j in range(len(link)):
    for k in range(1,2):
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        soup = BeautifulSoup(response.content) 
        for link3 in soup.findAll('a',{'class':'a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold'}):
            review_tag += link3.get_text()+'new_line'
            review_tag = review_tag.strip('\n')
```

```python
#user_id_link
user_id_link = ''
for j in range(len(link)):
    for k in range(1,3):
        response = Searchreviews(link[j]+'&pageNumber='+str(k))
        soup = BeautifulSoup(response.content) 
        for link4 in soup.findAll('a',{'class':'a-profile'})[2:]:
            user_id_link += link4.get('href')+ 'new_line'
            user_id_link = user_id_link.strip('\n')
```

```python
#conversion
def Convert(string): 
    li = list(string.split("new_line")) 
    return li
review_id_link = Convert(review_id_link)
review = Convert(review)
review_tag = Convert(review_tag)
user_id_link = Convert(user_id_link)
```

```python
df1 = pd.DataFrame(review_id_link,columns=['review_id_link'])
df1 = df1.iloc[:-1,:]
df2 = pd.DataFrame(review,columns=['review'])
df2 = df2.iloc[:-1,:]
df3 = pd.DataFrame(review_tag,columns=['review_tag'])
df3 = df3.iloc[:-1,:]
df4 = pd.DataFrame(user_id_link,columns=['userid_link'])
df4 = df4.iloc[:-1,:]
```

```python
review_scraper = {'name':profile_name,'ratings':ratings,'date':date}
review_scraper = pd.DataFrame.from_dict(review_scraper, orient='index')
review_scraper = review_scraper.transpose()
review_scraper['geography'] = 'USA'
pd.set_option('max_colwidth',800)
```

```python
#review_scraper.drop_duplicates
review_scraper.dropna(inplace=True)
```

```python
review_scraper = pd.concat([review_scraper,df1,df2,df3,df4], axis=1)
review_scraper.dropna(inplace=True)
```

```python
review_scraper.head()
```

```python
#user rating cleaning
def star_clean(df_col):
    new_column = []
    for i in df_col.map(lambda x: x.replace('out of 5 stars', '')):
        x = np.array(i, dtype= object)
        new_column.append(x) 
    return new_column
```

```python
#date conversion
def date_conv(d):
    d = str(d)
    x = re.search("on", d)
    d = d[int(x.end()+1):]
    return d
```

```python
review_scraper['ratings'] = star_clean(review_scraper.ratings)
review_scraper['date'] = review_scraper['date'].apply(date_conv)
```
